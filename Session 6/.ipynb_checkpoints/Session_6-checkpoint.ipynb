## 2. Advanced Visualizations

### 2.1 Pairwise Kernel Density Estimate (KDE) Plot `seaborn`

```python
# Filter the data for the year 2024
df_2024 = df[df['year'] == 2024]

# Apply log transformation to reduce skewness of variables
df_2024['log_population'] = np.log(df_2024['population'])
df_2024['log_gdp'] = np.log(df_2024['gdp_per_capita'])
df_2024['log_income'] = np.log(df_2024['daily_income'])
df_2024['log_mortality'] = np.log(df_2024['child_mortality'])

# Drop the original columns that have been transformed and any other unwanted columns
df_2024 = df_2024.drop(['country', 'year', 'population', 'gdp_per_capita',
                         'daily_income', 'child_mortality', 'co2_emission'], axis=1)

# Set the seaborn theme to a clean white background
sns.set_theme(style="white")

# Define a custom color palette for the 'continent' variable
color_dict = {
    "Asia": "#ffcc5c",         # orange
    "Europe": "#0392cf",       # blue
    "Africa": "#ff6f69",       # red
    "North America": "#88d8b0",# green
    "South America": "grey",   # grey
    "Oceania": "#854442"       # brown
}

# Create a PairGrid to plot pairwise relationships in the dataset
# The 'hue' parameter colors points by continent using the custom palette
g = sns.PairGrid(df_2024, hue="continent", palette=color_dict, diag_sharey=False)

# Map the upper triangle with scatter plots (with marker size 15)
g.map_upper(sns.scatterplot, s=15)

# Map the lower triangle with kernel density estimation (KDE) plots
g.map_lower(sns.kdeplot)

# Map the diagonal with KDE plots to show distribution of individual variables (line width 2)
g.map_diag(sns.kdeplot, lw=2)

# Add a legend to explain the color mapping for continents
g.add_legend()

# Display the final plot
plt.show()
```

### 2.2 Maps `matplot`

```python
import pandas as pd
import geopandas as gpd
from matplotlib.colors import LinearSegmentedColormap
import os
import matplotlib.pyplot as plt

# Set environment variable to restore/create the missing .shx file if needed
os.environ["SHAPE_RESTORE_SHX"] = "YES"

# -----------------------------
# Data Loading and Preparation
# -----------------------------

# Load the 2024 Gapminder data
df = pd.read_csv('Gapminder.csv')
df_2024 = df[df['year'] == 2024]
df_2024['country'] = df_2024['country'].replace('UK', 'United Kingdom')

# Load the world shapefile (ensure the correct path is provided)
world = gpd.read_file("ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp")
# Filter the shapefile to include only European countries
europe = world[world['CONTINENT'] == 'Europe']

# Merge the European geometry with the happiness dataset on country names
data = europe.merge(df_2024, left_on='NAME', right_on='country', how='left')

# -----------------------------
# Helper Functions
# -----------------------------

def create_gradient_colormap(colors):
    """
    Create a linear gradient colormap from a list of two colors.
    
    Parameters:
    colors (list): A list containing two color strings.
    
    Returns:
    cmap: A LinearSegmentedColormap object.
    """
    cmap = LinearSegmentedColormap.from_list("custom_gradient", colors, N=256)
    return cmap


def plot_map_on_ax(column, ax, cmap):
    """
    Plot a choropleth map for a given column on the specified axis using a provided colormap.
    
    Parameters:
    column (str): The column name in 'data' to visualize.
    ax (matplotlib.axes.Axes): The axis on which to plot the map.
    cmap: The colormap to use for the plot.
    """
    data.plot(
        column=column,          # Data column to visualize
        edgecolor='black',      # Country borders color
        linewidth=0.4,          # Thickness of the borders
        ax=ax,                  # Axis to plot on
        cmap=cmap,              # Colormap for the data values
        legend=True             # Display the colorbar legend
    )
    # Set the geographic limits to focus on Europe
    ax.set_xlim(-13.8, 40)
    ax.set_ylim(32, 72)
    ax.axis('off')              # Remove axis ticks and labels
    ax.set_title(column)

# -----------------------------
# Plotting Multiple Maps
# -----------------------------

# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 10))
# Flatten the 2D array of axes for easier iteration
axs = axs.flatten()

# List of columns (indicators) to plot
columns = [
    'life_expectancy',
    'daily_income',
    'child_mortality',
    'babies_per_woman'
]

# Define a dictionary with custom color gradients for each indicator
colors = {
    'life_expectancy': ['#c9ada7', '#4a4e69'],
    'daily_income': ['#fcbf49', '#d62828'],
    'child_mortality': ['#90e0ef', '#0077b6'],
    'babies_per_woman': ['#80ed99', '#38a3a5'],
}

# Iterate over each subplot axis and corresponding indicator column
for ax, column in zip(axs, columns):
    # Create a colormap for the current indicator based on its defined colors
    cmap = create_gradient_colormap(colors[column])
    # Plot the choropleth map for the current column on the given axis
    plot_map_on_ax(column=column, ax=ax, cmap=cmap)

# Adjust the layout to prevent overlapping elements and display the plot
plt.tight_layout()
plt.show()
```

### 2.3 Interactive Figures `plotly`

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

# Helper function to convert hex color to rgba string with given alpha
def hex_to_rgba(hex_color, alpha=1.0):
    hex_color = hex_color.lstrip('#')
    # Extract red, green, blue values and convert them to integers
    r, g, b = tuple(int(hex_color[i:i+2], 16) for i in (0, 2, 4))
    return f"rgba({r},{g},{b},{alpha})"

# Load the Gapminder dataset (update file path if needed)
df = pd.read_csv('Gapminder.csv')

# Filter the data to a single year (2024)
df_2024 = df[df.year == 2024]

# Create an income group based on daily income
bins = [0, 2, 8, 32, df_2024['daily_income'].max()]
labels = ['Less than $2', '$2-$8', '$8-$32', 'More than $32']
df_2024['income_group'] = pd.cut(df_2024['daily_income'], bins=bins, labels=labels)

# Aggregate data: total population by continent and income group
agg = df_2024.groupby(['continent', 'income_group'], as_index=False, observed=True)['population'].sum()

# Build a list of unique nodes: first continents, then income groups
continents = list(agg['continent'].unique())
income_groups = list(agg['income_group'].unique())
nodes = continents + income_groups

# Create a mapping from node name to index
node_indices = {node: i for i, node in enumerate(nodes)}

# Define custom colors for continents and income groups
continent_colors = {
    "Asia": "#ffcc5c",         # orange
    "Europe": "#0392cf",       # blue
    "Africa": "#ff6f69",       # red
    "North America": "#88d8b0",# green
    "South America": "grey",   # grey
    "Oceania": "#854442"       # brown
}

income_group_colors = {
    'Less than $2': "#03396c",            # darker
    '$2-$8': "#005b96",
    '$8-$32': "#6497b1",
    'More than $32': "#b3cde0"            # brighter
}

# Build the node color list: continents first then income groups
node_colors = []
for node in nodes:
    if node in continent_colors:
        node_colors.append(continent_colors[node])
    elif node in income_group_colors:
        node_colors.append(income_group_colors[node])
    else:
        node_colors.append("gray")  # fallback

# Create the sankey links: sources are continents, targets are income groups,
# and values are total populations.
source = []
target = []
value = []
link_colors = []  # We'll assign each link the color of the target income group (with transparency)
for _, row in agg.iterrows():
    source.append(node_indices[row['continent']])
    target.append(node_indices[row['income_group']])
    value.append(row['population'])
    # Get the base hex color for the income group and convert to a more transparent version (alpha=0.4)
    base_color = income_group_colors[row['income_group']]
    link_colors.append(hex_to_rgba(base_color, alpha=0.4))

# Create the interactive Sankey diagram using Plotly
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes,
        color=node_colors  # Use custom node colors
    ),
    link=dict(
        source=source,
        target=target,
        value=value,
        color=link_colors  # Each link gets the corresponding income group color with transparency
    )
)])

fig.update_layout(
    width=1000,   # set the width in pixels
    height=800,   # set the height in pixels
    title_text="Gapminder (2024): Population Flow from Continent to Income Group",
    font_size=10
)

fig.show()
```

### 2.4 Radar Plot `matplot`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# ------------------------------
# Data Preparation
# ------------------------------
# Load the Gapminder dataset (update file path if needed)
df = pd.read_csv('Gapminder.csv')

# Filter the data to a single year (2024)
df_2024 = df[df.year == 2024]
df_2024 = df_2024.drop(['year', 'country'], axis = 1).groupby('continent').mean()
grouped = df_2024[['life_expectancy', 'daily_income', 'child_mortality', 'babies_per_woman']].reset_index()

# Standardize columns
grouped['life_expectancy'] = (grouped['life_expectancy'] - grouped['life_expectancy'].min()) / (grouped['life_expectancy'].max() - grouped['life_expectancy'].min())
grouped['daily_income'] = (grouped['daily_income'] - grouped['daily_income'].min()) / (grouped['daily_income'].max() - grouped['daily_income'].min())
grouped['child_mortality'] = (grouped['child_mortality'] - grouped['child_mortality'].min()) / (grouped['child_mortality'].max() - grouped['child_mortality'].min())
grouped['babies_per_woman'] = (grouped['babies_per_woman'] - grouped['babies_per_woman'].min()) / (grouped['babies_per_woman'].max() - grouped['babies_per_woman'].min())

# Rename columns for clarity
grouped.rename(columns={'continent': 'continent',
                        'life_expectancy': 'Life Expectancy',
                        'daily_income': 'Daily Income',
                        'child_mortality': 'Child Mortality',
                        'babies_per_woman': 'Babies per Woman'}, inplace=True)

# ------------------------------
# Radar Chart Setup
# ------------------------------
# Define the variables (axes) for the radar plot
categories = ['Daily Income', 'Life Expectancy', 'Child Mortality', 'Babies per Woman']
N = len(categories)

# Calculate the angle for each axis (in radians)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]  # complete the loop

# ------------------------------
# Plotting
# ------------------------------
# Initialize the radar plot
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Place the first axis on top and go clockwise
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

# Draw one axis per variable + add variable labels
plt.xticks(angles[:-1], categories)

# Draw y-labels (customize as needed)
ax.set_rlabel_position(0)
plt.yticks([], [])  # Optionally remove radial ticks
plt.ylim(0, grouped[categories].max().max() * 1.1)

# Define a color palette for continents (modify colors as desired)
continent_colors = {
    "Asia": "#ffcc5c",         # orange
    "Europe": "#0392cf",       # blue
    "Africa": "#ff6f69",       # red
    "North America": "#88d8b0",# green
    "South America": "grey",   # grey
    "Oceania": "#854442"       # brown
}

# Plot a line for each continent
for idx, row in grouped.iterrows():
    values = row[categories].values.flatten().tolist()
    values += values[:1]  # complete the loop for the radar plot
    continent = row['continent']
    color = continent_colors.get(continent, 'gray')
    
    # Plot the outline
    ax.plot(angles, values, linewidth=2, linestyle='solid', label=continent, color=color)
    # Fill the area with some transparency
    ax.fill(angles, values, color=color, alpha=0.25)

# Add a legend outside the plot
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.title("Gapminder 2007: Average Indicators by Continent", y=1.08)

plt.show()
```
